```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rc('font', family='NanumBarunGothic')

# 엑셀 데이터 불러오기
file_path = '/content/drive/MyDrive/class_score.xlsx'
df = pd.read_excel('/content/drive/MyDrive/class_score.xlsx')

# 학급 내부 성적 비교
plt.figure(figsize=(10, 6))
sns.boxplot(x='class', y='score', data=df)
plt.title('Class Internal Score Comparison')
plt.show()

# 과목별 성적 추이
plt.figure(figsize=(10, 6))
sns.lineplot(x='subject', y='score', data=df)
plt.title('Subject Score Trend')
plt.show()

# 학급 전체 성적 분포
plt.figure(figsize=(10, 6))
sns.histplot(df['score'], kde=True)
plt.title('Total Score Distribution')
plt.show()
```
